**Matplotlib** is a Python library used for creating static, interactive, and animated visualizations. In machine learning, it helps visualize data, model performance (e.g., confusion matrices, ROC curves), and feature importance, making it easier to understand and interpret results.

### Pros of Matplotlib:

- Easy to use for basic plots.
- Highly customizable visualizations.
- Supports a wide range of plot types.
- Integrates well with other libraries (e.g., NumPy, Pandas).
- Extensive documentation and community support.

### Cons of Matplotlib:

- Steeper learning curve for complex visualizations.
- Can be slower for large datasets or highly interactive plots.
- Syntax can be verbose for simple tasks.
- Not as modern or interactive as some other libraries (e.g., Plotly, Seaborn).

```python
import matplotlib.pyplot as plt
%matplotlib inline
# %matplotlib inline is a magic command in Jupyter notebooks that allows you to display Matplotlib plots directly within the notebook, instead of opening them in a separate window.
```

```python
import numpy as np
```

```python
##Simple examples
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
#scatter plot
plt.scatter(x,y,c='g')#c is used to put color here g means green
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
#for saving graph use   plt.savefig('Test.png')
```

```python
y=x*x
```

```python
#line graph
plt.plot(x,y,'r*--',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2D Diagram')
```

```python
# Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'m*')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y)
plt.show()
```

```python
#sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
plt.show()
```

```python
#cos curve
x=np.arange(0,4*np.pi,0.1)
y=np.cos(x)
plt.title("cos wave form")
plt.plot(x,y)
plt.show()
```

```python
#in Sub plots
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
y_tan=np.tan(x)
y_cot=1/np.sin(x)
y_cosec=1/np.cos(x)
y_sec=1/np.tan(x)

plt.subplot(2,3,1)#2 rows 3 columns

plt.plot(x,y_sin,"r--")
plt.title("Sine")
          
plt.subplot(2,3,2)
plt.plot(x,y_cos,"g--")
plt.title("Cosine")

plt.subplot(2,3,3)
plt.plot(x,y_tan,"m--")
plt.title("Tan")

plt.subplot(2,3,4)
plt.plot(x,y_cot,"r--")
plt.title("Cot")

plt.subplot(2,3,5)
plt.plot(x,y_cosec,"g--")
plt.title("Cosec")

plt.subplot(2,3,6)
plt.plot(x,y_sec,"m--")
plt.title("Sec")

plt.tight_layout()
```

# BAR GRAPH

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

# Histogram

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=20)
#In Matplotlib, the bins parameter in a histogram specifies how the data 
#is grouped into intervals, or "bins." Each bin represents a range of data
#values, and the histogram shows how many data points fall into each bin.
plt.title("Histogram")
```

# Box Plot

```python
data=[np.random.normal(0,std,100)for std in range(1,4)]

#rectangular boxplot
plt.boxplot(data,vert=True,patch_artist=True)
#vert=True means show in vertical 
#and patch_artist=True means to fill color in BoxPlot
plt.show()
```

# Pie Chart

```python
#Data to plot
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) #explode 1st slice

plt.pie(sizes,explode=explode,labels=labels,colors=colors,
       autopct='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
