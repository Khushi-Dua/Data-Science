# Correlation with Heatmap

- A correlation heatmap visualizes the **correlation matrix of variables**.
- It shows the **strength and direction of relationships** (positive or negative) between variables.
- **Color coding**: Positive correlations are usually in one color (e.g., blue), negative correlations in another (e.g., red).
- Values range **from -1 (strong negative) to +1 (strong positive), with 0 indicating no correlation.**
- Helps **identify patterns, relationships, and potential multicollinearity in data**.
- Often used in **exploratory data analysis (EDA)**.

```python
df.corr(numeric_only=True)
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr(numeric_only=True))
#highly correlated is given with one color and less correlated with other color
```

# Here
- **Total Bill & Tip (Strong correlation)** → A lighter shade, meaning higher total bills tend to have higher tips.
- **Total Bill & Size (Moderate correlation)** → Slightly darker, meaning larger groups tend to spend more.
- **Tip & Size (Weakest correlation)** → The darkest shade, suggesting group size doesn’t strongly affect tip amount.

# Join Plot
A jointplot is a Seaborn visualization that shows the relationship between two variables along with their individual distributions. It combines a scatter plot (or other plot types) with marginal histograms or KDE plots. This helps visualize both the correlation between the variables and their individual distributions.

### Univariate Analysis


```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
#hex is fig inside jointplot
```

# Here
1.**Main Plot (Hexbin Plot)**

- Darker hexagons → High concentration of data points.
- Lighter hexagons → Fewer data points in those regions.
- Upward trend → Suggests a positive correlation between tip and total_bill, meaning higher total bills tend to have higher tips.
- Some isolated hexagons at higher values suggest outliers (few cases of very high bills and tips).

2️.**Histograms (Top & Right)**

- **Top Histogram (Tip Distribution)**
  - Most tips are between $2-$4, with fewer cases of very high tips.
- **Right Histogram (Total Bill Distribution)**
   - Most bills are around $10-$25, with a few reaching beyond $50.

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Here
This is another jointplot, but this time it uses a scatter plot with a regression line.


- 1. **Main Scatter Plot (Center)**
    - Each dot represents a data point (tip vs. total bill).
      -The positive slope of the regression line shows a positive correlation between tip and total_bill.
    - Some outliers (points far from the main cluster) are visible, especially at high total bill values.
- 2️. **Regression Line & Confidence Interval (Shaded Area)**

     - The blue line shows the best-fit linear trend.
     - The shaded area represents a confidence interval, indicating uncertainty in predictions.
    - The regression suggests that higher total bills generally lead to higher tips.
- 3️. **Histograms & KDE (Top & Right Sides)**

    - Tip distribution (Top Histogram) → Most tips are between $2-$4, with a few very high tips.
   - Total bill distribution (Right Histogram) → Most bills range from $10-$25, with fewer extreme values.

# Dist Plot
A distplot (distribution plot) in ML shows the histogram of a numerical feature along with a KDE (Kernel Density Estimate) curve. It helps visualize the data distribution, skewness, and outliers.

```python
sns.distplot(df['tip'])
```

# Here
This distplot shows the distribution of tip amounts:

- **Histogram (Bars)** → Shows frequency of different tip values.
-  **KDE Curve (Smooth Line)** → Estimates the probability density of tips.
- **Right-Skewed** → Most tips are between $2-$4, but some high tips (outliers) exist.

## Info Gained:

- Most tips are low (around $2-$4).
- Few high tip values (above $6).
- Might need scaling or outlier handling in ML models

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# Box Plot 
A boxplot in Seaborn displays the distribution of numerical data through quartiles, highlighting the median, interquartile range (IQR), and potential outliers. It helps visualize the spread and skewness of the data.

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

# Orient parameter
- The orient parameter in a boxplot controls the orientation of the plot.

- orient='v' (vertical) displays the boxes vertically (default).
- orient='h' (horizontal) displays the boxes horizontally.
- It determines whether the categorical variable is plotted on the x-axis (vertical) or y-axis (horizontal).

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other features
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# ViolinPlot
- A violinplot combines aspects of a boxplot and a kernel density plot. It shows the distribution of numerical data, displaying the density of the data at different values, along with the median and quartiles. This provides more insight into the distribution shape compared to a boxplot.

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
